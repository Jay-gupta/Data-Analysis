```python
is_pos = df['language'].str.contains('python')
python_lang = df[is_pos]
python_lang.shape
```

```python
is_pos = df['language'].str.contains('ct')
c_lang = df[is_pos]
c_lang.shape
```

```python
is_pos = df['language'].str.contains('cpp')
cpp_lang = df[is_pos]
cpp_lang.shape
```

```python
is_pos = df['language'].str.contains('java')
java_lang = df[is_pos]
java_lang.shape
```

```python
tokenized_python = []
for text in python_lang['program']:
    tokenized_python.append(nltk.word_tokenize(text))
```

```python
python = [(tokenized_python, 'python') for text in tokenized_python]
python
```

```python
tokenized_c = []
for text in c_lang['program']:
    tokenized_c.append(nltk.word_tokenize(text))
```

```python
tokenized_cpp = []
for text in cpp_lang['program']:
    tokenized_cpp.append(nltk.word_tokenize(text))
```

```python
tokenized_java = []
for text in java_lang['program']:
    tokenized_java.append(nltk.word_tokenize(text))
```
