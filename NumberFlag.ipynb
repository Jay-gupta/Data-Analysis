```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame()
x = pd.Series(np.arange(1,1001))
```

```python
y = [i for i in range(0,1000)]
```

```python
for i in range(len(y)):
    if(i%2==0):
        y[i] = str('red')
    else:
        y[i] = str('green')
```

```python
y = pd.Series(y)
```

```python
data['index'] = x
```

```python
data['color'] = y
```

```python
data.head()
```

```python
x_train = data.iloc[0:700,0]
y_train = data.iloc[0:700,1]
x_test = data.iloc[700:1000,0]
y_test = data.iloc[700:1000,1]
```

```python
x_train = np.array(x_train)
x_train = x_train.reshape(-1,1)

y_train = np.array(y_train)
y_train = y_train.reshape(-1,1)

x_test = np.array(x_test)
x_test = x_test.reshape(-1,1)

y_test = np.array(y_test)
y_test = y_test.reshape(-1,1)
```

```python
y_train
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
